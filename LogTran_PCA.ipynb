```python
import numpy as np
import pandas as pd
from composition_stats import clr
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
df = pd.read_csv('MOCK_DATA.csv')
df
```

```python
x = df.iloc[:,4:]
x
```

```python
y = df.iloc[:,:4]
```

```python
features = ['Au', 'Cu', 'Fe', 'Ag', 'Zn']
x_values = x.loc[:, features].values
x_values.shape
```

```python
x_norm = StandardScaler().fit_transform(x_values) # normalizing the features
x_norm.shape
```

```python
x_clr = clr(x_values)
x_clr.shape
```

```python
feat_cols = ['feature'+str(i) for i in range(x_norm.shape[1])]
feat_cols
```

```python
normalised_df = pd.DataFrame(x_norm,columns=feat_cols)
normalised_df
```

```python
transformed_df = pd.DataFrame(x_clr,columns=feat_cols)
transformed_df
```

```python
pca = PCA(n_components=3)
principalComponents = pca.fit_transform(x_norm)
```

```python
pca_clr = PCA(n_components=3)
principalComponents_clr = pca_clr.fit_transform(x_clr)
```

```python
principal_Df = pd.DataFrame(data = principalComponents
             , columns = ['principal component 1', 'principal component 2', 'principal component 3'])
principal_Df
```

```python
principal_trans_Df = pd.DataFrame(data = principalComponents_clr
             , columns = ['principal component 1', 'principal component 2', 'principal component 3'])
principal_trans_Df
```

```python
print('Explained variation per principal component: {}'.format(pca.explained_variance_ratio_))
```

```python
print('Explained variation per principal component for transformed data: {}'.format(pca_clr.explained_variance_ratio_))
```

```python
final_pca = y.join(principal_Df)
final_pca.to_csv('Final_Pca.csv')
final_pca
```

```python
final_trans_pca = y.join(principal_trans_Df)
final_trans_pca.to_csv('Final_Trans_Pca.csv')
final_trans_pca
```
